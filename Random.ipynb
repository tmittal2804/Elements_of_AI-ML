```python
import random
import matplotlib.pyplot as plt

# Generate random data for x and y
random.seed(42)
x = [random.randint(1, 50) for _ in range(100)]
y = [3 * xi + random.randint(-10, 10) for xi in x]

# Calculate means
x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)

# Calculate slope (theta_1) and intercept (theta_0)
num = sum((xi - x_mean) * (yi - y_mean) for xi, yi in zip(x, y))
deno = sum((xi - x_mean) ** 2 for xi in x)
theta_1 = num / deno
theta_0 = y_mean - theta_1 * x_mean

print(f"Slope (theta_1): {theta_1}")
print(f"Intercept (theta_0): {theta_0}")

# Define prediction function
def predict(x_values):
    return [theta_0 + theta_1 * xi for xi in x_values]

# Calculate predictions for original x values
y_pred = predict(x)
# print(f"preducted value:{y_pred}")

# Plot original data
plt.scatter(x, y, color='blue', label='Original Data')

# Plot regression line
x_line = range(min(x), max(x) + 1)
y_line = predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')
print("Tanya Mittal,590022823,07/02/2026,8:45 AM")

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
